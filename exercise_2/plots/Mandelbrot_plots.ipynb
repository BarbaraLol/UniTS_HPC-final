```python
# OMP scaling plot

import matplotlib.pyplot as plt
import numpy as np

# Number of threads and corresponding times (in seconds)
threads = np.array([1, 2, 4, 8, 12, 16, 20, 24])
times = np.array([63.56, 33.67, 18.69, 11.29, 8.58, 7.33, 6.70, 6.32])

# Parallelizable portion
P = 0.9

# Theoretical speedup using Amdahl's Law
theoretical_speedup = 1 / ((1 - P) + P / threads)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(threads, times, 'o-', color='blue', label='OMP scaling')
plt.plot(threads, times[0]/theoretical_speedup, 'g--', label='Theoretical speedup')  # Theoretical speedup scaled to the initial time

# Labeling the axes
plt.xlabel('Threads')
plt.ylabel('Time (s)')
plt.yscale('log')  # Use logarithmic scale for the y-axis
plt.title('OMP Scaling vs Theoretical Speedup')

# Adding a legend
plt.legend()

# Adding grid
plt.grid(True, which="both", ls="--")

# Display the plot
plt.show()
```

```python
# MPI scaling plot

import matplotlib.pyplot as plt
import numpy as np

# Number of processes and corresponding times (in seconds)
processes = np.array([2, 4, 8, 16, 24, 32, 48])
times = np.array([63.55, 22.38, 10.95, 6.63, 5.40, 4.79, 4.17])

# Parallelizable portion
P = 0.9

# Theoretical speedup using Amdahl's Law
theoretical_speedup = 1 / ((1 - P) + P / processes)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(processes, times, 'o-', color='blue', label='MPI scaling')
plt.plot(processes, times[0]/theoretical_speedup, 'g--', label='Theoretical speedup')  # Theoretical speedup scaled to the initial time

# Labeling the axes
plt.xlabel('Processes')
plt.ylabel('Time (s)')
plt.yscale('log')  # Use logarithmic scale for the y-axis
plt.title('MPI Scaling vs Theoretical Speedup')

# Adding a legend
plt.legend()

# Adding grid
plt.grid(True, which="both", ls="--")

# Display the plot
plt.show()
```

```python
# Strong MPI scaling plot

import matplotlib.pyplot as plt
import pandas as pd

# Data for strong scaling
strong_scaling_data = {
    'Processes': [2, 4, 8, 16, 24, 32, 48],
    'Time': [63.69, 14.20, 12.23, 10.04, 10.25, 7.30, 6.91]
}
df_strong = pd.DataFrame(strong_scaling_data)

# Theoretical Time for strong scaling
df_strong['Theoretical Time'] = df_strong['Time'].iloc[0] / df_strong['Processes']

# Plotting strong scaling
plt.figure(figsize=(10, 6))
plt.plot(df_strong['Processes'], df_strong['Time'], 'o-', label='MPI Strong Scaling')
plt.plot(df_strong['Processes'], df_strong['Theoretical Time'], 'g--', label='Theoretical Time')
plt.yscale('log')
plt.xlabel('Processes')
plt.ylabel('Time (s)')
plt.title('MPI Strong Scaling vs Theoretical Time')
plt.legend()
plt.grid(True)
plt.show()
```
